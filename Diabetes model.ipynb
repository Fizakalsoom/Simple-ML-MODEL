# **Diabetes model **

import dependies and libraries

```python
import pandas as pd #for handling and analyzing data in table format(dataset/data frame)
import numpy as np #for numerical oprations like working with arrays and math function
import matplotlib . pyplot as plt #for creation static plots and  graphs and charts
from sklearn.model_selection import train_test_split #for splitig data into traning and testing
from sklearn.metrics import accuracy_score,confusion_matrix#for evaluating the model perfomance
from sklearn.linear_model import LogisticRegression #best for binary classification
```

# **upload and import data set from drive to colab**

```python
from google.colab import drive
drive. mount('/content/drive')
dataset_path ='/content/drive/MyDrive/diabetes.csv'
```

# **Read dataset by using pandas library**

```python

dataset =pd.read_csv(dataset_path)
head =dataset.head(2) #head()print first 5 rows by default but we print only 2
print(head)
```

#numbers and columns in dataset

```python

shape =dataset.shape
print(shape) #dimension of dataset
```

# **name of columns**

```python

column_names = dataset.columns
print(column_names)
```

# **#check if is missing or not**

```python

mis_val = dataset.isnull().sum()
print(mis_val)
```

#divide dataset


```python

X =dataset[['Age', 'Gender', 'Polyuria', 'Polydipsia', 'sudden weight loss',
       'weakness', 'Polyphagia', 'Genital thrush', 'visual blurring',
       'Itching', 'Irritability', 'delayed healing', 'partial paresis',
       'muscle stiffness', 'Alopecia', 'Obesity']]
Y = dataset['class']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,random_state=42)
```

# **apply algorithm**

```python

model =LogisticRegression()
model.fit(X_train,Y_train)
Y_pred =model.predict(X_test)
```

# **Evaluate model**

```python
accuracy = accuracy_score(Y_test,Y_pred)
print(accuracy)
```

# **confusion matrix**

```python
conf_matrix =confusion_matrix(Y_test,Y_pred)
print(conf_matrix)
print(conf_matrix)
```

# **creat bar chart**

```python
labels = ['True Negative', 'False Positive', 'False Negative', 'True Positive']
values = [TN, FP, FN, TP]
TN, FP,FN,TP =conf_matrix.ravel()
```

# **plotting bar chart**
